```python
import pandas as pd
import sqlite3
conn = sqlite3.connect(r'C:\Users\patri\REPO_PRUEBA\ONLINE_DS_THEBRIDGE_Patricia_Diez\Sprint_08\challenge_06\sql-murder-mystery.db')
cursor = conn.cursor()
```

```python
def sql_query(query):
    cursor.execute(query)
    if cursor.description:  
        ans = cursor.fetchall()
        names = [description[0] for description in cursor.description]
        return pd.DataFrame(ans, columns=names)
    else:
        conn.commit() 
        return None
```

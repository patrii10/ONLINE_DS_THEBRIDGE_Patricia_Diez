## Ejercicio 1
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10

````python
primero = int(input("introduce un número: "))
segundo = int(input("otro porfa!: "))

def suma_valores(primero, segundo): 

    if (primero + segundo) == 10: 
        print("La suma de los dos igual a 10")
    elif (primero + segundo) > 10: 
        print("La suma de los dos es mayor que 10")
    elif (primero + segundo) < 10: 
        print("La suma de los dos es menor que 10")

suma_valores(primero, segundo)
````
